# Tokyo 2020 Olympic Games Data
Dataset sourced from https://www.kaggle.com/arjunprasadsarkhel/2021-olympics-in-tokyo

Throughout the many Olympic Games that has been held in modern times, the athletes who participates in the Games have always been a source of awe and inspiration. The carefully planned training and raw determination of each athlete is put on the ultimate test every 4 years for a chance at Olympic Gold and to push the very limits of human athleticism.

From the Olympic Games data gather from the recent Tokyo 2020 Olympics, we'll explore the different factors that could have affected the medal count of each participating nation, namely the number of participating athletes, the number of disciplines, and the number of coaches. For visualization, we'll use Seaborn and Matplotlib to help us digest the data in this notebook.

Primary Question: How does the number of participating athletes, number of disciplines, and number of coaches affect total medal count?

# Collecting and Loading Dataset

```python
# Import Dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating path
work_dir = os.getcwd()
main_dir = os.path.dirname(work_dir)
datasets_dir = main_dir + '/datasets'
```

```python
# Check files in datasets directory
dataset_list = os.listdir(datasets_dir)
dataset_list
```

```python
# Obtain direct path to datasets
file_names = []
for dataset in dataset_list:
    file_names.append(datasets_dir + "/" + dataset)
```

```python
# Loading datasets
df_list = []
for file in file_names:
    df_list.append(pd.read_excel(file))
```

# Basic Analysis of Dataset and Data Types

```python
# Naming loaded dataframes
coaches_df = df_list[0]
athletes_df = df_list[1]
gender_df = df_list[2]
teams_df = df_list[3]
medals_df = df_list[4]
```

```python
# Checking first few rows of all dataframes
print("Coaches DataFrame")
print(coaches_df.head())
print("-----------------------------------")
print("Athletes DataFrame")
print(athletes_df.head())
print("-----------------------------------")
print("Gender DataFrame")
print(gender_df.head())
print("-----------------------------------")
print("Teams DataFrame")
print(teams_df.head())
print("-----------------------------------")
print("Medals DataFrame")
print(medals_df.head())
```

```python
# Checking datatype in all DataFrames
print("Coaches DataFrame")
coaches_df.info()
print("-----------------------------------")
print("Athletes DataFrame")
athletes_df.info()
print("-----------------------------------")
print("Gender DataFrame")
gender_df.info()
print("-----------------------------------")
print("Teams DataFrame")
teams_df.info()
print("-----------------------------------")
print("Medals DataFrame")
medals_df.info()
```

Given we're interested in the data of each nation, we could use pandas' groupby function to help aggregate our data based on NOC, or Name of Country. The values would simply be the number of unique athletes each nation sends and discipline each nation participates in for the Olympic Games. From there, we could start exploring our data with exploratory data analysis.

# Data Cleaning

```python
# Grouping number of athletes and discipline by participating nation from athletes_df
athletes_bygroup_df = athletes_df.groupby("NOC").nunique()
athletes_bygroup_df
```

```python
# Renaming columns
athletes_bygroup_df = athletes_bygroup_df.rename({"Name":"Number of Athletes","Discipline":"Number of Disciplines"}, axis="columns")
athletes_bygroup_df
```

For our new dataframe, we'll also be interesting in coaching statistics and the number of medals.

```python
# Grouping number of coaches from coaches_df
coaches_bygroup_df = coaches_df.groupby("NOC").nunique()
coaches_bygroup_df
```

```python
# The value of coaches_bygroup_df["Name"] should be the number of coaches
# Combining athletes_bygroup_df with coaches_bygroup_df["Name"] using join with "NOC" as index
combined_df = pd.merge(athletes_bygroup_df, coaches_bygroup_df["Name"], how="left", on="NOC")
combined_df = combined_df.rename({"Name":"Number of Coaches"}, axis="columns")
```

```python
combined_df.info()
```

```python
combined_df["Number of Coaches"].unique()
```

```python
# Countries without coaches were be set as NaN during the merge
# For NaNs we'll replace with 0
combined_df.fillna(0, inplace=True)
combined_df
```

```python
combined_df.info()
```

```python
# Renaming Team/NOC with just NOC
medals_df = medals_df.rename({"Team/NOC":"NOC","Total":"Total Medals"},
                             axis="columns")
medals_df
```

For our analysis, we're not interested in medal rankings, just the raw number of medals. So, we'll move forward selecting out medal rankings, but keeping medal counts.

```python
# Selecting columns of interest from medals_df
columns_of_interest = ["NOC", "Gold", "Silver", "Bronze", "Total Medals"]
medal_count_df = medals_df[columns_of_interest].set_index("NOC")
medal_count_df
```

```python
medal_count_df.info()
```

```python
# Merging combined_df with medal_count_df
combined_df = pd.merge(combined_df, medal_count_df, how="left", on="NOC")
combined_df
```

```python
# Countries without medals were be set as NaN during the merge
combined_df.info()
```

```python
# Replace NaNs with 0
combined_df.fillna(0, inplace=True)
combined_df
```

```python
combined_df.info()
```

We could create several new columns in our dataframe which tells us the number of medals per number of athletes, number of coaches, or number of disciplines basis.

```python
# Deriving Number of Athletes per Number of Medals
combined_df["Gold per Number of Athletes"] = combined_df["Gold"] / combined_df["Number of Athletes"]
combined_df["Silver per Number of Athletes"] = combined_df["Silver"] / combined_df["Number of Athletes"]
combined_df["Bronze per Number of Athletes"] = combined_df["Bronze"] / combined_df["Number of Athletes"]
combined_df["Total Medals per Number of Athletes"] = combined_df["Total Medals"] / combined_df["Number of Athletes"]
```

```python
# Deriving Number of Coaches per Number of Medals
combined_df["Gold per Number of Coaches"] = combined_df["Gold"] / combined_df["Number of Coaches"]
combined_df["Silver per Number of Coaches"] = combined_df["Silver"] / combined_df["Number of Coaches"]
combined_df["Bronze per Number of Coaches"] = combined_df["Bronze"] / combined_df["Number of Coaches"]
combined_df["Total Medals per Number of Coaches"] = combined_df["Total Medals"] / combined_df["Number of Coaches"]
```

```python
# Deriving Number of Discipline per Number of Medals
combined_df["Gold per Number of Disciplines"] = combined_df["Gold"] / combined_df["Number of Disciplines"]
combined_df["Silver per Number of Disciplines"] = combined_df["Silver"] / combined_df["Number of Disciplines"]
combined_df["Bronze per Number of Disciplines"] = combined_df["Bronze"] / combined_df["Number of Disciplines"]
combined_df["Total Medals per Number of Disciplines"] = combined_df["Total Medals"] / combined_df["Number of Disciplines"]
```

```python
# Considering there are countries with 0 coaches, there will be infinities in our dataset
# There will also be countries that didn't won any medal with 0 coaches, these will end up be NaNs
combined_df["Total Medals per Number of Coaches"].unique()
```

```python
# We'll replace NaNs and infinities with 0
combined_df.replace(np.inf, 0, inplace=True)
combined_df.fillna(0, inplace=True)
combined_df
```

```python
# Checking data type
combined_df.info()
```

# Exploratory Data Analysis
Having cleaned our dataset, we could do dive a little deeper into our dataset.

```python
# Basic Statistics of dataset
combined_df.describe()
```

```python
tot_country = len(combined_df)
under_50_athl = len(combined_df[combined_df["Number of Athletes"] < 50])
under_15_disc = len(combined_df[combined_df["Number of Disciplines"] < 15])
no_coach = len(combined_df[combined_df["Number of Coaches"] == 0])
no_gold = len(combined_df[combined_df["Gold"] == 0])
no_silver = len(combined_df[combined_df["Silver"] == 0])
no_bronze = len(combined_df[combined_df["Bronze"] == 0])
no_medals = len(combined_df[combined_df["Total Medals"] == 0])
```

```python
columns = [tot_country, under_50_athl, under_15_disc, no_coach, 
           no_gold, no_silver, no_bronze, no_medals]
columns_names = ["Total Number of Countries", "Under 50 Athletes", "Under 15 Disciplines", "No Coach", 
                 "No Gold Medals", "No Silver Medals", "No Bronze Medals", "No Medals"]
f, ax = plt.subplots(figsize=(10, 10))
plt.barh(columns_names, columns)
plt.xlabel("Number of Countries")
for i, v in enumerate(columns):
    plt.text(x=v, y=i, s=str(v), va='center')
plt.title("Number of Countries vs. various conditions")
plt.show()
```

As shown from the simple statistics, we could see that a vast majority of nations has less than 50 athletes, participated in less than 15 disciplines, or has 1 coach or less. The same could be said for winning any medal from the Olympics.

```python
# Total medals vs. Number of Athletes
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
sns.scatterplot(x=combined_df["Number of Athletes"],
                y=combined_df['Total Medals'])
for line in range(0,combined_df.shape[0]):
    plt.text(combined_df["Number of Athletes"][line]+0.2,
             combined_df['Total Medals'][line],
             combined_df.index[line], horizontalalignment='left')
plt.title("Total Medals vs. Number of Athletes")
plt.show()
```

```python
# Pearson correlation between Number of Medals and Number of Athletes
medal_athl_corr = combined_df["Number of Athletes"].corr(combined_df['Total Medals'])
print("Pearson Correlation")
print("Medal vs. Number of Athletes: {:.2}".format(medal_athl_corr))
```

A large medal count is heavily correlated with a large athlete pool. This is readily apparent with the United States. Having won the most amount of medals with the most amount of athletes participating in the Games. However, it doesn't always guarentee a position at the podium. Japan, for example, has well over 500 participants. Australia has over 450 participants. However, both countries won less medals than China, Russia, and Great Britain, who's number of athletes are under 300. This indicates that the quality of the athlete themselves are important. This can be seen with Germany and China, who has a similar number of athletes, but China has nearly twice the amount of medals when compared to Germany during the Games.

```python
# Total medals vs. Number of Coaches
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
sns.scatterplot(x=combined_df["Number of Coaches"],
                y=combined_df['Total Medals'])
for line in range(0,combined_df.shape[0]):
    plt.text(combined_df["Number of Coaches"][line]+0.2,
             combined_df['Total Medals'][line],
             combined_df.index[line], horizontalalignment='left')
plt.title("Total Medals vs. Number of Coaches")
plt.show()
```

```python
# Pearson correlation between Number of Medals and Number of Athletes
medal_coach_corr = combined_df["Number of Coaches"].corr(combined_df['Total Medals'])
print("Pearson Correlation")
print("Medal vs. Number of Coaches: {:.2}".format(medal_coach_corr))
```

A large medal count is also correlated with a large coaching pool during the Olympics. Again, as shown with the United States. However, this too doesn't guarentee a position at the podium. Japan, for example, has 35 coaches for various teams at the events, but won less 60 medals. Australia and Spain, with 21 and 26 coaches, respectively, has won less than 50 and 20 overall medals, respectively. The examples of China, Russia, and Great Britain indicates that the quality of coaching also matters to winning medals.

```python
# Total medals vs. Number of Disciplines
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
sns.scatterplot(x=combined_df["Number of Disciplines"],
                y=combined_df['Total Medals'])
for line in range(0,combined_df.shape[0]):
    plt.text(combined_df["Number of Disciplines"][line]+0.2,
             combined_df['Total Medals'][line],
             combined_df.index[line], horizontalalignment='left')
plt.title("Total Medals vs. Number of Disciplines")
plt.show()
```

```python
# Pearson correlation between Number of Medals and Number of Discipline
medal_coach_corr = combined_df["Number of Disciplines"].corr(combined_df['Total Medals'])
print("Pearson Correlation")
print("Medal vs. Number of Disciplines: {:.2}".format(medal_coach_corr))
```

```python
combined_df["Number of Disciplines"]["Cuba"]
```

Similar to a large number of athletes and a large number of coaches, participating in a large number of disciplines (or sports) also correlates to a large total medal count. This indicates that having more teams across multiple sports would increase the overall chances of the country of winning a medal. However, similar to before, this does not guarentee a medal for that country. Chile is a prime example of this with a grand total of 0 overall medals, despite having participated in 25 separate sports. The same could be said for Mexico and Argentina, where each total medal count are less than 5, despite taking part in over 30 separate disciplines. Cuba, on the other hand, has won 15 overall medals in the 15 discipline they participated in.
