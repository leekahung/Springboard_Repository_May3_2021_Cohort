These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
# Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017
url = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X/data.json?start_date=2017-01-01&end_date=2017-12-31&api_key=' + API_KEY
r = requests.get(url)
#print(r)
```

```python
# Convert the returned JSON object into a Python dictionary
json_data = r.json()
#print(json_data)
```

```python
# Calculate what the highest and lowest opening prices were for the stock in this period.
data_2017_columns = json_data['dataset_data']['column_names']
data_2017 = json_data['dataset_data']['data']
#print(data_2017_columns)
#print(data_2017)

data_2017_open = []

for daily_data in data_2017:
    price_open = daily_data[data_2017_columns.index('Open')]
    if price_open != None:
        data_2017_open.append(price_open)

#print(data_2017_open)
highest_open = max(data_2017_open)
lowest_open = min(data_2017_open)
print("Highest opening price for this stock during this period was", round(highest_open, 2))
print("Lowest opening price for this stock during this period was", round(lowest_open, 2))
```

```python
# What was the largest change in any one day (based on High and Low price)?
data_2017_daily_price_change = []

for daily_data in data_2017:
    price_high = daily_data[data_2017_columns.index('High')]
    price_low = daily_data[data_2017_columns.index('Low')]
    price_change = price_high - price_low
    data_2017_daily_price_change.append(price_change)

#print(data_2017_price_change)
largest_change = max(data_2017_daily_price_change)
print("The largest one day change was", round(largest_change, 2))
```

```python
# What was the largest change between any two days (based on Closing Price)?
data_2017_close = []

for daily_data in data_2017:
    price_close = daily_data[data_2017_columns.index('Close')]
    data_2017_close.append(price_close)

#print(data_2017_close)

data_2017_2day_price_change = []

for i in range(1, len(data_2017_close)):
    price_change = data_2017_close[i] - data_2017_close[i - 1]
    data_2017_2day_price_change.append(price_change)
    
#print(data_2017_two_day_price_change)
largest_2day_change = max(data_2017_2day_price_change)
print("The largest two day change was", round(largest_2day_change, 2))
```

```python
# What was the average daily trading volume during this year?
data_2017_trading_volume = []

for daily_data in data_2017:
    data_2017_trading_volume.append(daily_data[data_2017_columns.index('Traded Volume')])
    
#print(data_2017_trading_volume)
avg_trade_vol_2017 = statistics.mean(data_2017_trading_volume)
print("The average daily trading volume during 2017 was", round(avg_trade_vol_2017))
```

```python
# What was the median trading volume during this year?
median_trade_vol_2017 = statistics.median(data_2017_trading_volume)
print("The median daily trading volume during 2017 was", round(median_trade_vol_2017))
```
