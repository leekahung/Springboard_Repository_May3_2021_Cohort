Q10. Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
# Q10
query_Q10 = '''
WITH guest AS (
    SELECT
        f.name,
        b.facid,
        SUM(b.slots) * f.guestcost AS guest_revenue
    FROM Bookings as b
    INNER JOIN Facilities as f
    USING(facid)
    WHERE b.memid = 0
    GROUP BY facid
),
members AS (
    SELECT
        f.name,
        b.facid,
        SUM(b.slots) * f.membercost AS member_revenue
    FROM Bookings as b
    INNER JOIN Facilities as f
    USING(facid)
    WHERE b.memid != 0
    GROUP BY facid
)

SELECT
    f.name,
    guest.guest_revenue + members.member_revenue AS revenue
FROM Bookings
INNER JOIN Facilities as f
USING(facid)
INNER JOIN guest
USING(facid)
INNER JOIN members
USING(facid)
GROUP BY facid
HAVING revenue < 1000
ORDER BY revenue DESC;'''
```

```python
df_Q10 = pd.read_sql_query(query_Q10, connection)
```

Q11. Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
# Q11
query_Q11 = '''
SELECT
    a.surname || ', ' || a.firstname AS surname_firstname,
    b.surname || ', ' || b.firstname AS recommededby_name
FROM Members AS a
LEFT JOIN Members AS b
ON a.recommendedby = b.memid
WHERE a.recommendedby != ''; -- apparently empty string '' is not null'''
```

```python
df_Q11 = pd.read_sql_query(query_Q11, connection)
```

```python
df_Q11
```

Q12. Find the facilities with their usage by member, but not guests

```python
# Q12
query_Q12 = '''
SELECT
    f.name,
    ROUND(SUM(b.slots) * 1.0 / (SELECT SUM(slots) FROM Bookings) * 100, 2) AS member_percent_usage
FROM Bookings as b
INNER JOIN Facilities as f
USING(facid)
WHERE b.memid != 0
GROUP BY facid;'''
```

```python
df_Q12 = pd.read_sql_query(query_Q12, connection)
```

```python
df_Q12
```

Q13. Find the facilities usage by month, but not guests

```python
# Q13
query_Q13 = '''
WITH july AS (
    SELECT
        f.name,
        b.facid,
        ROUND(SUM(b.slots) * 1.0 / (SELECT SUM(slots) FROM Bookings) * 100, 2) AS jul_percent_usage
    FROM Bookings as b
    INNER JOIN Facilities as f
    USING(facid)
    WHERE strftime('%m', b.starttime) = '07'
    GROUP BY facid
),    
august AS (
    SELECT
        f.name,
        b.facid,
        ROUND(SUM(b.slots) * 1.0 / (SELECT SUM(slots) FROM Bookings) * 100, 2) AS aug_percent_usage
    FROM Bookings as b
    INNER JOIN Facilities as f
    USING(facid)
    WHERE strftime('%m', b.starttime) = '08'
    GROUP BY facid
),
september AS (
    SELECT
        f.name,
        b.facid,
        ROUND(SUM(b.slots) * 1.0 / (SELECT SUM(slots) FROM Bookings) * 100, 2) AS sept_percent_usage
    FROM Bookings as b
    INNER JOIN Facilities as f
    USING(facid)
    WHERE strftime('%m', b.starttime) = '09'
    GROUP BY facid
)
 
SELECT
    f.name,
    july.jul_percent_usage,
    august.aug_percent_usage,
    september.sept_percent_usage
FROM Bookings
INNER JOIN Facilities as f
USING(facid)
INNER JOIN july
USING(facid)
INNER JOIN august
USING(facid)
INNER JOIN september
USING(facid)
GROUP BY facid;'''
```

```python
df_Q13 = pd.read_sql_query(query_Q13, connection)
```
